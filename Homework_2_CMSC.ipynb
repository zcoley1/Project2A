```python
#We are going to run hypothesis tests on two non-numerical values (with the help of AI)
import pandas as pd
from scipy.stats import chi2_contingency
import seaborn as sns
import matplotlib.pyplot as plt

#dropping missing values... again
df_clean = df[["scenario_4", "scenario_9"]].dropna()

#merge responses
politics_map = {
    "Strongly liberal": "Liberal",
    "Mildly liberal": "Liberal",
    "Strongly conservative": "Conservative",
    "Mildly conservative": "Conservative",
    "Neutral": "Neutral",
    "Don't know / It's complicated": "Unsure"
}

fetish_map = {
    "Does not describe me at all": "Barely Describes Me",
    "Slightly describes me": "Barely Describes Me",
    "Moderately describes me": "Describes Me",
    "Describes me fairly well": "Describes Me"
}

#Insert new map into responses
df_clean["scenario_politics"] = df_clean["scenario_4"].map(politics_map)
df_clean["scenario_fetish"] = df_clean["scenario_9"].map(fetish_map)
```

```python
#Using a chi-square hypothesis test because our data is not numerical
contingency_table = pd.crosstab(df_clean['scenario_politics'], df_clean['scenario_fetish'])
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(contingency_table)
```

```python
print(f"Chi-squared statistic: {chi2}")
print(f"P-value: {p}")
```

```python
#Using AI to make a stacked bar chart to represent our findings
ct = pd.crosstab(df_clean["scenario_fetish"], df_clean['scenario_politics'])
ct_proportions = ct.div(ct.sum(axis=1), axis=0)
ct_proportions.plot(kind='bar',
                    stacked=True,
                    figsize=(10, 6),
                    color=['#e74c3c', '#3498db', '#95a5a6', '#f1c40f'])
plt.title('Tolerance of Fetish by Political Lean', fontsize=14)
plt.ylabel('Proportion (0.0 to 1.0)')
plt.xlabel('Fetish Tolerance')
plt.xticks(rotation=0)
plt.legend(title='Political Leaning', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.savefig('fetish_politics_chart.png')
```

```python
#Running Our second hypothesis test
import pandas as pd
from scipy.stats import chi2_contingency
import seaborn as sns
import matplotlib.pyplot as plt

#dropping the useless data
df_clean = df[["scenario_6", "scenario_26"]].dropna()

#merge responses
virginity_map = {
    "Describes me fairly well" : "Describes me",
    "Describes me extremely well" : "Describes me",
    "Moderately describes me" : "Neutral",
    "Slightly describes me" : "Barely Describes Me",
    "Does not describe me at all" : "Barely Describes Me"
}

#inserting new map
df_clean["scenario_virginity"] = df_clean["scenario_26"].map(virginity_map)
```

```python
#Running the Chi-Square Test
contingency_table = pd.crosstab(df_clean['scenario_virginity'], df_clean['scenario_6'])
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(contingency_table)
```

```python
print(f"Chi-squared statistic: {chi2}")
print(f"P-value: {p}")
```

```python
#Using AI to make a pretty graph
ct = pd.crosstab(df_clean["scenario_virginity"], df_clean['scenario_6'])
ct_proportions = ct.div(ct.sum(axis=1), axis=0)
ct_proportions.plot(kind='bar',
                    stacked=True,
                    figsize=(10, 6),
                    color=['#fca9fc', '#a2a7f5', '#ed8082', '#d681f7'])
plt.title('Virgin Morals by Gender', fontsize=14)
plt.ylabel('Proportion (0.0 to 1.0)')
plt.xlabel('Virgin Morals')
plt.xticks(rotation=0)
plt.legend(title='Gender', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
```
